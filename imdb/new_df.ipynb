```python
import pandas as pd
```

```python
df = pd.read_csv('imdb_movies_new.csv')
df.head(3)
```

## Cleaning the dataset

```python
df = df.drop('Cast', axis=1)
df.head(3)
```

```python
df['Actors'] = df['Actors'].str.split(',').str[0]
df['Actress'] = df['Actress'].str.split(',').str[0]
```

```python
df['Actors'][0:5]
```

```python
df['Actress'][0:5]
```

## Dataset after being cleaned

```python
df
```

## Removing Null values

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df = df.rename(columns={'Movie title': 'Movie_title'})
```

## Saving the dataset to a sqlite database

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///imdb.db')
df.to_sql('movies', engine)
```

## Descriptive analysis of the data

#### What is the average rating of the movies in the table?

```python
avg_rating = df['Rating'].mean()
avg_rating
```

#### Who is the director of the movie with the highest gross revenue?


```python
max_gross_row = df.loc[df['Gross'].idxmax()]

# extract the director name from the row
director_name = max_gross_row['Director']
director_name
```

#### How many movies in the table have a duration of less than 100 minutes?


```python
# filter the DataFrame to include only rows with duration less than 200 minutes
short_movies = df.loc[df['Duration'] < 100]

# count the number of rows in the filtered DataFrame using the 'count' function
num_short_movies = short_movies['Movie_title'].count()
num_short_movies
```

#### Which movie has the highest rating?

#### How many different genres are there in the table and what are they?


```python
genre_series = df['Genre']

# split the genres by ',' to create a list of individual genres
genre_list = []
for genres in genre_series:
    genre_list.extend(genres.split(', '))

# convert the list to a set to remove duplicates and count the number of unique genres
num_genres = len(set(genre_list))
```

#### What is the total gross revenue for all the movies in the table?


```python
total_gross = df['Gross'].sum()
total_gross
```

#### Which director has the most movies in the table and how many movies did they direct?


```python
# group the DataFrame by directors and count the number of movies for each director
director_counts = df.groupby('Director')['Movie_title'].count().reset_index()

# sort the director counts in descending order
sorted_director_counts = director_counts.sort_values(by='Movie_title', ascending=False)

# take the top 10 directors with the most movies
top_10_directors = sorted_director_counts.head(10)
top_10_directors
```

#### How many movies did actress Zoe Saldana appear in?


```python
# use the 'str.contains' method to find all rows where 'Actress' contains 'Zoe Saldana'
zoe_movies = df[df['Actress'].str.contains('Zoe Saldana')]

# count the number of rows to find the number of movies that Zoe Saldana appeared in
num_movies = zoe_movies.shape[0]

# print the result
print("Zoe Saldana appeared in", num_movies, "movies.")
```

```python
# use boolean indexing to filter the rows where 'Actress' is 'Zoe Saldana'
saldana_movies = df[df['Actress'] == 'Zoe Saldana']

# extract the relevant columns from the filtered DataFrame
movie_titles = saldana_movies['Movie_title']
actress_names = saldana_movies['Actress']

# create a new DataFrame with the movie titles and actress names
saldana_df = pd.DataFrame({'Movie title': movie_titles, 'Actress': actress_names})
saldana_df
```

#### Were the movies that actor Stephen Lang appeared in a success?


```python
# use the 'str.contains' method to find all rows where 'Actors' contains 'Stephen Lang'
lang_movies = df[df['Actors'].str.contains('Stephen Lang')]

# calculate the average rating for the movies that Stephen Lang appeared in using the 'mean' function
avg_rating = lang_movies['Rating'].mean()

# check if the average rating is above a certain threshold, indicating success
if avg_rating > 7:
    print("The movies that Stephen Lang appeared in were a success.")
else:
    print("The movies that Stephen Lang appeared in were not a success.")
```

#### How many movies did director James Cameron make that grossed over $300,000?


```python
# use boolean indexing to filter the rows where 'Director' is 'James Cameron' and 'Gross' is over $300,000
cameron_movies = df[(df['Director'] == 'James Cameron') & (df['Gross'] > 300000)]

# count the number of rows to find the number of movies that meet the criteria
num_movies = cameron_movies.shape[0]

# print the result
print("James Cameron made", num_movies, "movies that grossed over $300,000.")
```

```python
# use boolean indexing to filter the rows where 'Director' is 'James Cameron' and 'Gross' is over $300,000
cameron_movies = df[(df['Director'] == 'James Cameron') & (df['Gross'] > 300000)]

# extract the relevant columns from the filtered DataFrame
movie_titles = cameron_movies['Movie_title']
gross_amounts = cameron_movies['Gross']

# create a new DataFrame with the movie titles and gross amounts
cameron_df = pd.DataFrame({'Movie_title': movie_titles, 'Gross': gross_amounts})

# format the 'Gross' column as currency
cameron_df['Gross'] = cameron_df['Gross'].apply(lambda x: '${:,.2f}'.format(x))
cameron_df
```

#### Is there any relationship between the movie's genre and its box office success?

```python
genre_gross = df.groupby('Genre')['Gross'].mean().sort_values(ascending=False)
genre_gross
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# create a horizontal bar chart using Seaborn
sns.set(style='whitegrid')
plt.figure(figsize=(20, 14))
ax = sns.barplot(x='Gross', y='Genre', data=genre_gross.reset_index(), color='steelblue')

# set the x-axis label and increase fontsize
ax.set_xlabel('Gross', fontsize=24)

# set the y-axis label and increase fontsize
ax.set_ylabel('Genre', fontsize=24)

# set the title and increase fontsize
ax.set_title('Average Gross by Genre', fontsize=28)

# rotate the x-axis labels
plt.xticks(rotation=45)

# display the chart
plt.show()

```
